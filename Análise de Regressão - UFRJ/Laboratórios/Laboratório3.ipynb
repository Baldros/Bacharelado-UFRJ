**Instalando Packages**

```r
# Desinstalar o pacote xfun
remove.packages("xfun")
```

```r
# Listar as versões disponíveis do xfun no CRAN
available_versions <- available.packages()["xfun", "Version"];available_versions
```

```r
desired_version <- available_versions

# Verificar se a versão desejada está disponível
if (desired_version %in% available_versions) {
  # Instalar a versão desejada do xfun
  install.packages("xfun", version = desired_version)
} else {
  # Se a versão desejada não estiver disponível, informar ao usuário
  message("A versão ", desired_version, " do pacote xfun não está disponível.")
}
```

```r
# Verificar a versão do pacote xfun
packageVersion("xfun")
```
